```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = {
    'Age': [22, 34, 45, 23, 25, 34, 32, 41, 36, 29, 40, 27, 23, 31, 34, 30, 27, 22, 24, 26],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Male',
               'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male']
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8, 6))
```

```python
sns.countplot(x='Gender', data=df, palette='Set2')
```

```python
plt.title('Gender Distribution')
```

```python
plt.xlabel('Gender')
```

```python
plt.ylabel('Count')
```
